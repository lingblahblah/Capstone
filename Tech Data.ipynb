```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
files = os.listdir(path + '/Tech')                                # change direction
files_xls = [f for f in files if f[-4:] == 'xlsx']                # a list of all excel files
prod_152 = [f for f in files if f[:3] == '152']                   # a list of excel files for product 152
g1_1400rpm = [f for f in files if f[-15:] == '1g-1400rpm.xlsx']   # a list of all excel file under 1g-1400rpm  
```

```python
'''
For product 152, 
scatter plots of its heights and volumes verse time, 
under measurement 1g-1400rpm
'''

data_152_1_h = pd.read_excel(path + '/Tech/152 000044 1g-1400rpm.xlsx', 'Height - Raw data')
data_152_1_s = pd.read_excel(path + '/Tech/152 000044 1g-1400rpm.xlsx', 'Structure - Raw data')

data_152_1_h = data_152_1_h[np.isfinite(data_152_1_h['Foam Volume [mL]'])]
data_152_1_s = data_152_1_s[np.isfinite(data_152_1_s['SD MBA [m]'])]

fig = plt.figure(figsize = (18, 15))
st = fig.suptitle('Produce 152 \nUnder Measurement Method 1g-1400rpm', fontsize = 20)

ax1 = fig.add_subplot(2,2,1)
line1, = ax1.plot(data_152_1_h['Time [s]'], data_152_1_h['Foam Height [mm]'], 'o', label = 'Foam', alpha = 0.5)
line2, = ax1.plot(data_152_1_h['Time [s]'], data_152_1_h['Liquid Height [mm]'], 'o', label = 'Liquid', alpha = 0.5)
line3, = ax1.plot(data_152_1_h['Time [s]'], data_152_1_h['Total Height [mm]'], 'o', label = 'Foam + Liquid', alpha = 0.5)
ax1.legend(loc='upper right')
ax1.set_title('Height')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Height [mm]')
ax1.set_xlim([150,180])          # many duplicates after 200s

ax2 = fig.add_subplot(2,2,2)
line1, = ax2.plot(data_152_1_h['Time [s]'], data_152_1_h['Foam Volume [mL]'], 'o', label = 'Foam', alpha = 0.5)
line2, = ax2.plot(data_152_1_h['Time [s]'], data_152_1_h['Liquid Volume [mL]'], 'o', label = 'Liquid', alpha = 0.5)
line3, = ax2.plot(data_152_1_h['Time [s]'], data_152_1_h['Total Volume [mL]'], 'o', label = 'Foam + Liquid', alpha = 0.5)
ax2.legend(loc='upper right')
ax2.set_title('Volumn')
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Volume [mL]')
ax2.set_xlim([150,180])

ax3 = fig.add_subplot(2,2,3)
line1, = ax3.plot(data_152_1_s['Time [s]'], data_152_1_s['BC/mm'], 'o', label = 'Bubble Count', alpha = 0.5)
ax3.set_title('Bubble Count')
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('BC/mm')
ax3.set_xlim([150,180])

ax4 = fig.add_subplot(2,2,4)
line1, = ax4.plot(data_152_1_s['Time [s]'], data_152_1_s['MBA [m]'], 'o', label = 'Liquid', alpha = 0.5)
ax4.set_title('Mean Bubble Area')
ax4.set_xlabel('Time [s]')
ax4.set_ylabel('MBA [m]')
ax4.set_xlim([150,180])

plt.show()

#fig.savefig('prod_152_1.png')
#plt.close(fig)
```
